#### OK / Récupérer TOP Album par année depuis 1994

```python
# url_base = 'https://snepmusique.com/les-tops/le-top-de-lannee/top-albums-annee/?annee=2024'

# il faudra boucler sur le dernier nombre
# url_year = 'https://snepmusique.com/les-tops/le-top-de-lannee/top-albums-annee/?annee='+{loop}

# création des listes
list_rang_album = []
list_artist_album = []
list_album_album= []
list_id_album = []

# boucle de 1994 à 2024
for loop in range(1994, 2025):
    
    # print(int(loop))
    
    # url sur laquelle on boucle pour les années
    url_year_album = 'https://snepmusique.com/les-tops/le-top-de-lannee/top-albums-annee/?annee='+str(loop)
    # print(url_year)
    
    # request sur l'url
    html_a = requests.get(url_year_album)
    
    # créer la soupe du lien album
    soup_a = BeautifulSoup(html_a.text, 'html.parser')
    
    # récupérer le 'bloc' contenant les top recherchés (artiste, titre/album, rang)
    charts_a = soup_a.find_all('div', {'class' :'item'})

    for chart_a in charts_a :
        # print(chart)

        # récupérer le rang et l'ajouter à la liste
        list_rang_album.append((chart_a.find('div', {'class' : 'rang'})).text.strip())
        
        # récupérer le nom de l'artiste et l'ajouter à la liste
        list_album_album.append((chart_a.find('div', {'class' : 'titre'})).text.strip())
        
        # récupérer le nom de l'album et l'ajouter à la liste
        list_artist_album.append((chart_a.find('div', {'class' : 'artiste'})).text.strip())
    
        # récupérer l'id spotify quand elle existe (seulement depuis 2023) et l'ajouter à la liste
        try:
            id_spot_album = (re.findall("(?<=\/album/)(.*)",
                    chart_a.find('a', {'class' : 'icon-spotify'})['href']))
            list_id_album.append(id_spot_album[0])
        except:
            list_id_album.append('no id')

    # rassembler les données dans un DF
    df_top_album =pd.DataFrame(zip(list_rang_album, list_artist_album, list_album_album, list_id_album),
                        columns=['rang', 'artiste', 'album', 'id_spotify'])

df_top_album
```
